```python
import pandas as pd
import matplotlib.pyplot as plt

# create a sample DataFrame
df = pd.DataFrame({'date': ['2021-01-01', '2021-02-01', '2022-01-01', '2022-02-01'],
                   'sales': [100, 200, 150, 250]})

# convert the date column to datetime format
df['date'] = pd.to_datetime(df['date'])

# group the sales data by year and sum the values
sales_by_year = df.groupby(df['date'].dt.year)['sales'].sum()

# plot a line chart of the sales data
plt.plot(sales_by_year.index, sales_by_year.values)

# add labels and title to the chart
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Yearly Sales Data')

# display the chart
plt.show()
```
