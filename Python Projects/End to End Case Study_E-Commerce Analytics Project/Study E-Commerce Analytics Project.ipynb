## b. Understanding how many new customers acquired every month


```python
customers_acquired = pd.merge(left=customer, right=orders, how='inner', on='customer_id')
```

```python
customers_acquired.head()
```

```python
customers_acquired['order_yr_mnth'] =  customers_acquired["order_purchase_timestamp"].dt.strftime('%Y-%m')
```

```python
customers_acquired_a = customers_acquired.loc[:, ['customer_unique_id', 'order_yr_mnth']] 
```

```python
customers_acquired_a['customer_unique_id'].duplicated().sum()
```

```python
customers_acquired_a.drop_duplicates(subset='customer_unique_id', inplace=True)
```

```python
customers_acquired_a['customer_unique_id'].duplicated().sum()
```

```python
customers_acquired_a = customers_acquired_a.groupby(['order_yr_mnth'])[['customer_unique_id']].count().reset_index()
```

```python
customers_acquired_a
```

## c. Understand the retention of customers on month on month basis

```python
customer_retain = customers_acquired.loc[:, ['customer_unique_id', 'order_yr_mnth']] 
```

```python
customer_retain = (customer_retain.groupby('order_yr_mnth')[['customer_unique_id']].count()).reset_index()
```

```python
customer_retain = customer_retain.rename(columns={'customer_unique_id' : 'customer_unique_id_2'})
```

```python
customer_retain = pd.merge(left=customer_retain, right=customers_acquired_a, how='inner', on= 'order_yr_mnth')
```

```python
Geo_order_Table = pd.merge(Geo_LOCATION,sellers,left_on='geolocation_zip_code_prefix',right_on='seller_zip_code_prefix')

Ord_Pro_Table = pd.merge(order_items, products)

Ord_Ord_Pay = pd.merge(orders, order_payments ,on='order_id')

Major_Table_1 = pd.merge(Geo_order_Table, Ord_Pro_Table ,on='seller_id')

Major_Table_1 = pd.merge(Major_Table_1, Ord_Ord_Pay ,on='order_id')

Major_Table_1 = pd.merge(Major_Table_1, order_review_ratings ,on='order_id')

Major_Table = Major_Table_1.loc[:, [
    'customer_id',
    'seller_id',
    'order_id',
    'product_id',
    'order_purchase_timestamp',
    'product_category_name',
    'payment_type',
    'price',
    'payment_value',
    'geolocation_state',
    'review_score'
]]
```

```python
Major_Table.drop_duplicates(inplace=True)
```

```python
# Product count
Prod_cat_count = Major_Table.groupby('product_category_name').agg({
    'product_id':
    'count',
    'price':
    'sum'
}).rename(columns={'product_id':'product_count',
                  'price':'Total_price'})

Prod_cat_count.sort_values('product_count', ascending=False)
```

```python
# Product Count by state 
Prod_cat_count_by_state = Major_Table.groupby('geolocation_state').agg({
    'product_id':
    'count',
    'price':
    'sum'
}).rename(columns={'product_id':'product_count',
                  'price':'Total_price'})

Prod_cat_count_by_state
```

```python
# Revenue_by_state
Revenue_by_state = Major_Table.groupby('geolocation_state').agg({
    'price': 'sum'
}).rename(columns={
    'price': 'Total_Revenue'
}).sort_values(by='Total_Revenue', ascending=False)
Revenue_by_state
```

```python
# Customer Count by satate
customer_count_by_state = Major_Table.groupby('geolocation_state')[[
    'customer_id'
]].nunique().rename(columns={'customer_id': 'customer_id_count'})

customer_count_by_state
```

```python
Major_Table['order_purchase_Month'] = Major_Table['order_purchase_timestamp'].dt.strftime('%B')

Major_Table['order_purchase_week'] = Major_Table['order_purchase_timestamp'].dt.isocalendar().week 

Major_Table['order_purchase_Hour'] = Major_Table['order_purchase_timestamp'].dt.strftime('%HH')

```

```python
Major_Table['week_name'] = Major_Table['order_purchase_timestamp'].dt.strftime('%A')
```

```python
 Major_Table.groupby('order_purchase_Month').agg({
    'product_id':
    'count',
    'price':
    'sum'
}).rename(columns={'product_id':'product_count',
                  'price':'Total_price'})
```

```python
# Most Active Hours for purchase
(Major_Table.groupby('order_purchase_Hour').agg({
    'price':
    'sum'
}).rename(columns={'price':'Total_price'})).sort_values(by='Total_price',ascending=False)
```

```python
# Most Active Day for purchase
(Major_Table.groupby('week_name').agg({
    'price':
    'sum'
}).rename(columns={'price':'Total_price'})).sort_values(by='Total_price',ascending=False)
```

```python
 Major_Table.groupby('order_purchase_week').agg({
    'product_id':
    'count',
    'price':
    'sum'
}).rename(columns={'product_id':'product_count',
                  'price':'Total_price'})
```

# Q2. Performing Customers/sellers Segmentation
a. Divide the customers into groups based on the revenue generated


```python
Major_Table_1['Revenue'] = Major_Table_1['price']+Major_Table_1['freight_value']

cust_rev = Major_Table_1.groupby('customer_id')[['Revenue']].sum().sort_values(by='Revenue', ascending=False)
```

```python
cust_rev['cust_rev_group'] = pd.cut(
    cust_rev.Revenue,
    bins=np.arange(0, 15000, 1000),
    labels=[
        '0-1000', '1000-2000', '2000-3000', '3000-4000', '4000-5000',
        '5000-6000', '6000-7000', '7000-8000', '8000-9000', '9000-10000',
        '10000-11000', '11000-12000', '12000-13000', '13000-14000'
    ])

cust_rev 
```

## b. Divide the sellers into groups based on the revenue generated 

```python
seller_rev = Major_Table.groupby('seller_id')[['price']].sum().sort_values(
    by='price', ascending=False)
```

```python
seller_rev['cust_rev_group'] = pd.cut(seller_rev.price,
                                      bins=np.arange(0, 300000, 50000),
                                      labels=[
                                          '0-50000', '50000-100000',
                                          '100000-150000', '150000-200000',
                                          '200000-250000'
                                      ])
seller_rev
```
